```python
import pandas as pd
X_train = pd.read_csv(r"/workspaces/regularized-linera-regression/data/interim/X_train.csv")
X_test = pd.read_csv(r"/workspaces/regularized-linera-regression/data/interim/X_test.csv")
X_train
```

```python
X_test
```

```python
y_train = pd.read_csv(r"/workspaces/regularized-linera-regression/data/interim/y_train.csv")
y_test = pd.read_csv(r"/workspaces/regularized-linera-regression/data/interim/y_test.csv")
y_test
```

```python
y_train
```

```python
from pickle import dump, load
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.feature_selection import f_regression, SelectKBest

mse_list = []

for p in [1, 0.8, 0.7, 0.6, 0.5]:
    selection_model = SelectKBest(f_regression, k = int(len(X_train.columns) * p))
    selection_model.fit(X_train, y_train)

    ix = selection_model.get_support()
    train_sel = pd.DataFrame(selection_model.transform(X_train), columns = X_train.columns.values[ix])

    

    model = Lasso(alpha = 3, max_iter = 300, random_state = 14)
    model.fit(train_sel, y_train)
    y_pred = model.predict(train_sel)
    mse = mean_squared_error(y_train, y_pred)
    mse_list.append(mse)


best_mse = mse_list.index(max(mse_list))

```

```python
best_index=int(len(X_train.columns)*[1, 0.8, 0.7, 0.6, 0.5][best_mse])
best_index
```

```python
dump(selection_model, open(r"/workspaces/regularized-linera-regression/models/selection_model.pk", "wb"))
```

```python
from sklearn.feature_selection import f_regression, SelectKBest

selection_model = SelectKBest(f_regression, k = best_index)
selection_model.fit(X_train, y_train)
ix = selection_model.get_support()
X_train_sel = pd.DataFrame(selection_model.transform(X_train), columns = X_train.columns.values[ix])
X_test_sel = pd.DataFrame(selection_model.transform(X_test), columns = X_test.columns.values[ix])




X_train_sel
```
